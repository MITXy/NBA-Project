```python
fig, ax = plt.subplots(figsize=(6,5))
sns.barplot(y="player", x='points_per_game', data=df3_sort, color="yellow")
plt.title('Points Per Game for Point Guards in NBA History', color="b")
plt.xlabel("Points", color="b")
plt.ylabel("Player", color="b")

# Add labels inside boxes
for i, box in enumerate(ax.containers):
    for j, bar in enumerate(box):
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + bar.get_height() / 2, 
                str(round(bar.get_width(), 1)), color='b', fontsize=10, ha='center', va='center')
        
plt.show()
```

```python
from PIL import Image

# Load and resize image
img = Image.open("C:/Users/DELL/OneDrive/Documents/Data/steph-cat.png")
img = img.resize((300, 300))

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot bar chart
sns.barplot(y="player", x='points_per_game', data=df3_sort, color="yellow", ax=ax1)
ax1.set_title('Points Per Game for Point Guards in NBA History', color="b")
ax1.set_xlabel("Points", color="b")
ax1.set_ylabel("Player", color="b")

# Add labels inside boxes
for i, box in enumerate(ax1.containers):
    for j, bar in enumerate(box):
        ax1.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + bar.get_height() / 2, 
                 str(round(bar.get_width(), 1)), color='b', fontsize=10, ha='center', va='center')

# Plot image
ax2.imshow(img)
ax2.axis('off')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(y="player", x='points_per_game', data=df3_sort, color="blue", ax=ax)
ax.set_title('Points Per Game for Point Guards in NBA History', color="black", fontsize=16)
ax.set_xlabel("Points", color="black", fontsize=12)
ax.set_ylabel("Player", color="black", fontsize=12)
ax.set_facecolor('gold')

# Add labels inside the bars
for i, box in enumerate(ax.containers):
    for j, rect in enumerate(box.patches):
        box_name = box.get_label()[j]
        box_value = str(round(rect.get_width(), 1))
        ax.text(rect.get_x() + rect.get_width()/2, rect.get_y() + rect.get_height()/2, f"{box_name}\n{box_value}",
                 ha='center', va='center', fontsize=10, color='black')

# Load and display the image
img = plt.imread("C:/Users/DELL/OneDrive/Documents/Data/steph-cat.png")
ax2 = fig.add_axes([0.7, 0.1, 0.2, 0.2])
ax2.imshow(img)
ax2.axis('off')

plt.show()
```
